```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('flying-etiquette.csv')
print(df.shape)
df.head(5)
```

```python
graph1 = cleandf['How often do you travel by plane?'].value_counts()
graph1.plot(kind='barh')
```

```python
graph2 = cleandf['Who should have control over the window shade?'].value_counts()
graph2.plot(kind='barh')
```
